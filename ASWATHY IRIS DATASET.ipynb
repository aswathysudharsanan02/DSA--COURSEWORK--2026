```python
import pandas as pd

# Read the Iris dataset
df = pd.read_excel("iris.xls")

print("Dataset successfully read into Python")
df.head(10)
```

**DATASET PREVIEW**

```python
df.head(10).style.set_properties(
    **{
        'background-color': '#f0f8ff',
        'color': 'black',
        'border': '1px solid black'
    }
).set_table_styles([
    {
        'selector': 'th',
        'props': [
            ('background-color', '#4CAF50'),
            ('color', 'white'),
            ('font-weight', 'bold')
        ]
    }
]).set_caption("Iris Dataset")
```

**2. DISPLAY COLUMNS IN DATASET**

```python
print("Columns in the dataset:")
print(df.columns)
```

**3. CALCULATE MEAN OF EACH COLUMN**

```python
df.mean(numeric_only=True).to_frame(name="Mean Value").style.set_properties(
    **{
        'background-color': '#e6f7ff',
        'color': 'black',
        'border': '1px solid #ccc'
    }
).set_table_styles([
    {
        'selector': 'th',
        'props': [
            ('background-color', '#007acc'),
            ('color', 'white')
        ]
    }
])
```

**4. CHECKING NULL VALUES**

```python
df.isnull().sum().to_frame(name="Missing Values").style.set_properties(
    **{
        'background-color': '#fff3cd',
        'color': 'black',
        'border': '1px solid #ccc'
    }
).set_table_styles([
    {
        'selector': 'th',
        'props': [
            ('background-color', '#ff9800'),
            ('color', 'white')
        ]
    }
])
```

**5. DATA VISUALIZATIONS**

   **(a) Scatter Plot (Petal Features)**

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    x="PL",
    y="PW",
    hue="Classification",
    data=df,
    palette="Set1",
    s=90
)

plt.title("Petal Length (PL) vs Petal Width (PW)")
plt.show()
```

**(b) Histogram (Sepal Length Distribution)**

```python
plt.figure(figsize=(7,5))

plt.hist(
    df["SL"],
    bins=15,
    color="#6A5ACD",
    edgecolor="black",
    alpha=0.85
)

plt.title("Distribution of Sepal Length (SL)")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

**(c) Box Plot (Feature Comparison)**

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df.drop("Classification", axis=1),
    palette="Set1"
)

plt.title("Feature Distribution of Iris Dataset")
plt.show()
```

**(d) Pairplot**

```python
sns.pairplot(df, hue="Classification", palette="husl")
plt.show()
```
