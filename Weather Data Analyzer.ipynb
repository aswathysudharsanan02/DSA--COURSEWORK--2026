```python
# Weather Data Analyzer

# Taking input
n = int(input("Enter number of days: "))

temperatures = []
humidities = []

# Input data
for i in range(n):
    temp = float(input("Enter temperature for day " + str(i+1) + ": "))
    hum = float(input("Enter humidity for day " + str(i+1) + ": "))
    
    temperatures.append(temp)
    humidities.append(hum)

# Function to calculate average
def calculate_average(data):
    total = sum(data)
    avg = total / len(data)
    return avg

# Function to determine weather condition
def determine_weather(avg_temp, avg_humidity):
    if avg_temp > 30 and avg_humidity < 50:
        return "Sunny"
    elif avg_humidity > 70:
        return "Rainy"
    elif avg_temp < 20:
        return "Cold"
    else:
        return "Moderate"

# Function to display results
def display_results():
    avg_temp = calculate_average(temperatures)
    avg_humidity = calculate_average(humidities)
    condition = determine_weather(avg_temp, avg_humidity)

    print("\n--- Weather Report ---")
    print("Average Temperature:", avg_temp)
    print("Average Humidity:", avg_humidity)
    print("Weather Condition:", condition)

# Run program
display_results()
```
